# ANN

### Data Exploration--Intuition

```python
# installed Keras, tensorflow, and theano. 

## Read the data. 

import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import display


data=pd.read_csv('Churn_Modelling.csv')
labels=data['Exited']
features=data.drop(['RowNumber','CustomerId','Surname','Exited'],axis = 1)
print ("Churning Customers dataset has {} samples with {} features each.".format(*features.shape))
display(features.head(n=1))
display(labels.head(3))
display(data.head(n=3))
```

```python
display(features.describe())
churned_customers=data.loc[labels==0]

print("Total number of Cusomers Who have left the bank: {}".format(len(churned_customers)))
```

### Dealing with Categorical Variables. 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
features_cat=features.select_dtypes(include=['object'])
features['Geography']=labelencoder_X_1.fit_transform(features_cat['Geography'])
display(features.head())



             
             
```

```python
labelencoder_X_2 = LabelEncoder()
features['Gender']=labelencoder_X_2.fit_transform(features_cat['Gender'])
display(features.head())
```

```python
onehotencoder = OneHotEncoder(categorical_features = [1])
features = onehotencoder.fit_transform(features).toarray()
display(features[:5,:])
features=features[:,1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.2, random_state = 42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Neural Netork Build

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier=Sequential()

classifier.add(Dense(activation="relu",input_dim=11, units=6, kernel_initializer="uniform"))
classifier.add(Dense(activation="relu",units=6, kernel_initializer="uniform"))
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

classifier.fit(X_train,y_train,batch_size=10,nb_epoch=100)

y_pred=classifier.predict(X_test)





```

```python
display(y_pred)
```

### Thresold-- Predictions

```python
y_pred=(y_pred>0.5)
display(y_pred[10:20])
```

### Confusion Matrix


```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
display(cm)
```
